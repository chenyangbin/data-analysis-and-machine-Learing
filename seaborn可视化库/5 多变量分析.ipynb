### 点图描述数据的变化的差异
* 描述自变量和因变量的变化趋势 类似于折线图

```python
titanic.head()
```

```python
sns.pointplot(x="Sex", y="Survived", hue="Pclass", data=titanic)
```

### 点图指定颜色以及线形进行区分
* 线颜色参数 palett=(参量, 颜色)
* 变量标记形状参数 markers=[" ", " "] 标记不同hue *^&
* 线形状参数 linestyle=[" ", " "] 标记不同的线 - --
* 设置不同的绘图方式 kind=" " box 箱线图

```python
sns.pointplot(x="Pclass", y="Survived", hue="Sex", data=titanic,
              palette={"male":"g", "female":"m"}, # 颜色
              markers=["o", "^"], # 标记形状
              linestyles=["-", "--"]
              )
```
