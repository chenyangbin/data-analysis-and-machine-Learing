```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(color_codes=True)

np.random.seed(sum(map(ord, "regression")))

tips = sns.load_dataset("tips")  # 导入内置数据集合, tips 消费集合, 需要翻墙
tips.head()
```

### 使用 regplot() 和 lmplot() 绘制线性回归关系
* 二者均可
* regplot 比较简单实用
* mlplot 复杂但是功能多


```python
sns.regplot(x="total_bill", y="tip", data=tips)
#sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
# size和tip的会馆关系
sns.regplot(x="size", y="tip", data=tips)
```

### 离散数据 类别等的回归需要加上抖动处理,更加趋于连续化
* x/y_jitter

```python

#size 为离散值2 3 4 5等 不适合做回归,
# 要对数据加入小幅度的抖动
sns.regplot(data=tips, x="size", y="tip", x_jitter=.05) # size 进行上下0.05 抖动
```
