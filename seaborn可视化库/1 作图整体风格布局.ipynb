### seaborn库是基于matlabplot封装的画图库，内置众多模板。使用方便快捷


```python
# 使用
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
%matplotlib inline

# 使用plot进行绘图的时候，参数的配置比较麻烦
# 因此，使用具有模板的seaborn比较高效
def sinplot(flip=1):
    x = np.linspace(0,14,100)  # 14个的区间找100个点
    for i in  range(1,7): # 画7条线
        plt.plot(x, np.sin(x + i*5) * (7-i) * flip)
sns.set()
sinplot()
```

### seaborn中的5种主题风格

### seaborn主题示例

```python
# 白色格子主题的seaborn
sns.set_style("whitegrid")
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

### 黑色主题 dark

```python
sns.set_style("dark")
sinplot()
```

### 白色主题 white

```python
# 白色背景无刻度线
sns.set_style("white")
sinplot()
```

### ticks主题

```python
# ticks主题 白色主题， x，y轴中， 上下左右具有刻度线
sns.set_style("ticks")
sinplot()
```

### 去除上以及右侧的线段

```python
sinplot()
sns.despine() # 去除轴线画风
```

### 对不同的风格加以参数

### 指定图像距离轴线的距离

```python
sns.violinplot(data)
sns.despine(offset=10)
```

### 轴线的保留与否 .despine(left=, right=)

```python
sns.set_style("whitegrid")  # 白色横隔主题
sns.boxplot(data=data, palette="deep")  
sns.despine(right=True, left=True)  # 默认false 保留
```

### 指定子图的风格 with域指定批量子图风格

```python
with sns.axes_style("darkgrid"): # 域中的图像的风格都是一样de
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)

```

### 设置画图区域的大小, 字体大小, 线的粗细

```python
sns,set() # 图默认值
sns.set_context('paper')  #指定画图的大小
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns,set() # 图默认值
sns.set_context('talk')  #指定画图的大小
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns,set() # 图默认值
sns.set_context('notebook')  #指定画图的大小
plt.figure(figsize=(8, 6))
sinplot()
```

### 自定义设置图表中的字体的大小

```python
sns,set() # 图默认值
sns.set_context('talk', font_scale=1.5, rc={'lines.linewidth':2.5}, )  #指定画图的大小
plt.figure(figsize=(8, 6))
sinplot()
```
