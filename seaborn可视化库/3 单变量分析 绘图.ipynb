### 取到数据以后
* 简单了解数据的分布状况, 单变量单特征分析

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)  # 使用默认颜色编码
np.random.seed(sum(map(ord, "distribution")))
```

```python
# 初始数据显示
x = np.random.normal(size=100) # 随机高斯数据
sns.distplot(x, kde=False) # 直方图 简单了解特征分布
```

### 源数据切分为均等分,查看均等分的数据的分布 bins参数

```python
sns.distplot(x, bins=20)
```

### 使用某个统计指标 查看数据分布的状况 fit 参数

```python
sns.distplot(x, bins=20, fit=stats.gamma) # 显示 gama, 以及高斯分布线
```

### 根据指定均值以及协方差生成数据 np.random.multivariate_normal(, , )

```python
mean, cov = [0,1], [(1, .5),(.5, 1 )]
data = np.random.multivariate_normal(mean, cov, 1000) # 根据均值以及协方差生成200组数据
df = pd.DataFrame(data, columns=["x", "y"])
df
```

### 特征和特征之间的关系 jointplot
* np.jointplot() 变量自身的分布 变量之间的关系分布 计算相关系数
* 散点图 
* 颜色深表示附近分布的点多

```python
#sns.jointplot(x="x", y="y", data=df)
sns.jointplot(x="x", y="y", data=df)

sns.jointplot(x="x", y="y", data=df, stat_func=stats.pearsonr, kind='hex',color='g' )
```
