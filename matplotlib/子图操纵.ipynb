### plot的多图 子图的绘制

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10)) # 绘制底层画布 figsize:配置图尺寸
ax1 = fig.add_subplot(4,3,1)  # 4x3 的第一幅
ax2 = fig.add_subplot(4,3,2)  # 4x3 的第二幅
ax3 = fig.add_subplot(4,3,3)
ax6 = fig.add_subplot(4,3,6)  # 4x3 的第6幅图
ax1.plot(np.random.randint(1,5,5), np.arange(5)) # 对ax1进行绘图
ax2.plot(np.arange(10)*3, np.arange(10))  # 绘制ax2图
plt.show()
```

### 一图多线绘制


```python
import pandas as pd
import matplotlib.pyplot as plt
unrate = pd.read_csv('unrate.csv')  # 月度失业率
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate['MONTH'] = unrate['DATE'].dt.month # 构建month属性值
fig = plt.figure(figsize=(6,3))
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='yellow')
plt.show()
 
```

```python
# 快捷绘制带图例的1图多线
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black'] # 颜色列表
for i in range(5): # 循环作图
    start_index = i*12 # 起始索引
    end_index = (i+1) * 12 # 结束索引
    subset = unrate[start_index:end_index] # 索引范围
    label = str(1948+i) # 标签
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
plt.legend(loc='best') # 标签位置 best为自动配置可放置在左上, 右下等位置
plt.title("duo line ")
plt.xlabel("month")
plt.ylabel("unrate")
plt.show()
```
